```python
import pandas as pd
import glob
import numpy as np
import python_speech_features as mfcc
from scipy.io.wavfile import read
import random
import regex as re
```

```python
LABELS = {
    "en": 0,
    "ur": 1,
    "ue": 2
}
NUM_CLASSES = 3
NUM_FEATURES = 13

RANDOM_SEED = 50

TRAIN_DATASET_GLOB = "sampled_dataset/train/*"
TEST_DATASET_GLOB = "sampled_dataset/test/*"

SHUMAIL_DATASET = "shumail2/*"

DATASPLIT = 0.8
```

```python
def get_MFCC(audio, sr):
    features = mfcc.mfcc(audio, sr, 0.025, 0.01, 13, appendEnergy=True)
    return np.mean(features, axis=0)
```

```python
data = []
labels = []

for path in glob.glob("Recordings/*"):
    label = path.split("\\")[1]
    label = re.split(r"-|_", label)[0]
    y = np.zeros((NUM_CLASSES))
    try: 
        labels.append(LABELS[label])

        sr, audio = read(path)
        features = get_MFCC(audio, sr)
        
        data.append(features)
    except KeyError: 
        print(KeyError, path)
        break

sample_index = list(range(0, len(data)))
random.shuffle(sample_index)

```
